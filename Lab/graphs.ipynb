```python
import matplotlib.pyplot as plt
import numpy as np

# Create sample data
'''x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)'''
plt.figure(figsize=(10, 5))  # Create the figure

# Plot 1 (First row, first column)
plt.subplot(1, 2, 1)  # (rows, columns, plot number)
plt.plot(x, main.times_counting, color='b', label='Counting')
plt.title('Counting Sort')
plt.legend()

# Plot 2 (First row, second column)
plt.subplot(1, 2, 2)
plt.plot(x, main.times_pigeon, color='r', linestyle='--', label='Pidgeon')
plt.title('Pidgeon Sort')
plt.legend()

plt.tight_layout()  # Adjust spacing to prevent overlap
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 8))  # 2x2 grid

# Top-left: Sine
ax[0, 0].plot(x, main.times_pigeon, color='blue')
ax[0, 0].set_title('Sine Function')

# Top-right: Cosine
ax[0, 1].plot(x, main.times_counting, color='orange')
ax[0, 1].set_title('Cosine Function')

# Bottom-left: Exponential
ax[1, 0].plot(x, main.times_quick_random, color='green')
ax[1, 0].set_title('Exponential Function')

# Bottom-right: Logarithmic
ax[1, 1].plot(x, main.times_quick_three, color='red')
ax[1, 1].set_title('Logarithmic Function')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))

x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), label='sin(x)', color='blue')
plt.plot(x, np.cos(x), label='cos(x)', linestyle='--', color='orange')
plt.plot(x, np.tan(x), label='tan(x)', linestyle=':', color='green')

plt.title("Multiple Functions in One Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.ylim(-2, 2)  # Limit the y-axis
plt.legend()
plt.grid(True)
plt.show()
```
